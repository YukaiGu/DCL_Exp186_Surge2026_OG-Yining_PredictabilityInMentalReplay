##### ========================================================================================================================
### 1. IMPORT LIBRARIES
##### ========================================================================================================================

```python
import pandas as pd
import numpy as np
import json
from pathlib import Path
import statsmodels.formula.api as smf
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import scipy
from scipy.ndimage import gaussian_filter1d
from scipy.signal import find_peaks
import ast
```

##### ========================================================================================================================
### 2. Define Data Path & Import Data
##### ========================================================================================================================

```python
data_dir = Path("Utility/Data/PData5")

files = sorted(data_dir.glob("*.csv"))

print("Files found:")
for f in files:
    print(f.name)

dfs = []

for f in files:
    df = pd.read_csv(f)
    df["source_file"] = f.name
    dfs.append(df)

data = pd.concat(dfs, ignore_index=True)
```

##### ========================================================================================================================
### 3. Data Cleaning
##### ========================================================================================================================

```python
# Keep only real segmentation video trials
Pilot5_data = data[data["trial_kind"] == "segmentation_video"].copy()

Pilot5_data.head()

#Remove Demo Trail
Pilot5_data = Pilot5_data[
    Pilot5_data["predictability"] != "Demo"].copy()

# Keep only the varaibles I need
keep_cols = [
    "PROLIFIC_PID",
    "randomized_pair_order",
    "stimulus_name",
    "predictability",
    "boundary_times_sec",
    "boundary_count",
    "block",
    "global_position_in_block",
    "video_attention_failure_type"
]

Pilot5_data = Pilot5_data[keep_cols].copy()
Pilot5_data = Pilot5_data.reset_index(drop=True)


#Remove Attention Violation Trails
Pilot5_data = Pilot5_data[
    Pilot5_data["video_attention_failure_type"].isna()
].copy()
```

##### ========================================================================================================================
### 4. Regression on NoB in Predictability Conditions
##### ========================================================================================================================

```python
# Make sure predictability is categorical
Pilot5_data["predictability"] = (Pilot5_data["predictability"].astype("category"))

# Linear regression
model = smf.ols("boundary_count ~ predictability",data=Pilot5_data).fit()

print(model.summary())
```

```python
# Plot for Regression

# Fit regression
Pilot5_data["predictability_num"] = Pilot5_data["predictability"].map({"Predictable": 0, "Unpredictable": 1 }).astype(float)

Pilot5_data["boundary_count"] = pd.to_numeric(Pilot5_data["boundary_count"], errors="coerce")

Pilot5_data = Pilot5_data.dropna(subset=["predictability_num", "boundary_count"]).copy()

model = smf.ols("boundary_count ~ predictability_num", data=Pilot5_data).fit()

x_pred = np.linspace(0, 1, 100)

pred_df = pd.DataFrame({"predictability_num": x_pred})

pred_summary = model.get_prediction(pred_df).summary_frame(alpha=0.05)

plt.figure(figsize=(7,5))

x_jitter = (Pilot5_data["predictability_num"] + np.random.normal(0, 0.03, len(Pilot5_data)))

plt.scatter(x_jitter, Pilot5_data["boundary_count"], alpha=0.3)
plt.plot(x_pred, pred_summary["mean"], linewidth=3)
plt.fill_between(x_pred, pred_summary["mean_ci_lower"], pred_summary["mean_ci_upper"], alpha=0.2 )
plt.xticks([0, 1], ["Predictable", "Unpredictable"])
plt.xlabel("Condition")
plt.ylabel("Boundary Count")
plt.title("Linear Regression of Boundary Count on Predictability")
plt.show()
```

##### ========================================================================================================================
### 5. Boundary Consensus
##### ========================================================================================================================

```python
# Video Duration Dictionary
video_durations = {
    "baking": 28.75,
    "balcony": 28.00,
    "bank": 19.11,
    "bathroom": 35.11,
    "beach": 26.75,
    "bedding": 34.68,
    "bike": 29.31,
    "car": 39.44,
    "cleaning": 30.00, 
    "cereal": 37.71,
    "fireplace": 30.99,
    "football": 36.00,
    "gym": 29.31,
    "lamp": 29.52,
    "laundry": 30.84,
    "mouse": 29.76,
    "music": 38.12,
    "painting": 24.55,
    "party": 35.17,
    "poster": 23.53,
    "printer": 44.05,
    "record": 19.11,
    "shopping": 11.79,
    "skateboard": 24.12,
    "suitcase": 27.15,
    "sunbathing": 40.12,
    "tea": 31.00, 
    "tennis": 29.76,
    "walking": 21.52,
    "whiteboard": 36.54,
}
```

```python
# Unpredictable manipulation timestamps
unpredictable_timestamps = {
    "baking": 12.00,
    "balcony": 19.00,
    "bank": 11.00,
    "bathroom": 17.00,
    "beach": 18.00,
    "bedding": 11.00,
    "bike": 12.00,
    "car": 18.00,
    "cleaning": 8.00,
    "cereal": 17.00,
    "fireplace": 16.80,
    "football": 12.00,
    "gym": 9.00,
    "lamp": 17.00,
    "laundry": 7.00,
    "mouse": 18.00,
    "music": 14.00,
    "painting": 9.00,
    "party": 15.00,
    "poster": 14.50,
    "printer": 19.00,
    "record": 9.00,
    "shopping": 6.00,
    "skateboard": 11.00,
    "suitcase": 12.00,
    "sunbathing": 16.00,
    "tea": 15.00,
    "tennis": 14.50,
    "walking": 7.70,
    "whiteboard": 14.00,
}

# Convert to rectangle windows
unpredictable_windows = {
    vid: (max(0, t - 0.5), t + 2.5)
    for vid, t in unpredictable_timestamps.items()
}
```

```python
# Parse the Boundary List
def parse_boundary_times(x):
    if pd.isna(x):
        return []
    if isinstance(x, list):
        return x
    try:
        return ast.literal_eval(x)
    except:
        return []

Pilot5_data["boundary_times_list"] = (Pilot5_data["boundary_times_sec"].apply(parse_boundary_times))


# Parameters 
# Smooth_sigma = Standard Deviation --> If SS = 1; BS = 0.5, then 1+- (1*0.5) would be detected
# Peak_distance is in bins size distance
# Peak_Prominence --> Threshold of the peak minus the valley
BIN_SIZE = 0.5
SMOOTH_SIGMA = 0.85

PEAK_PROMINENCE = 0.4
PEAK_DISTANCE = 1
```

```python
# Mean number of boundaries for each video × condition
mean_nob = (
    Pilot5_data
    .groupby(["stimulus_name", "predictability"])["boundary_count"]
    .mean()
)

# BUILD CONSENSUS TIMELINES BY VIDEO × CONDITION

all_video_results = {}
group_cols = ["stimulus_name", "predictability"]

for (stimulus_name, predictability), df_vid in Pilot5_data.groupby(group_cols):
    duration = video_durations[stimulus_name]
    all_boundaries = []

    #Aggregate all participant timestamps
    for boundary_list in df_vid["boundary_times_list"]:
        all_boundaries.extend(boundary_list)

    bins = np.arange(0,duration + BIN_SIZE,BIN_SIZE)
    counts, edges = np.histogram(all_boundaries,bins=bins)
    time_axis = edges[:-1]
    n_trials = len(df_vid)

    #Normalize it under the density function of 1
    consensus = counts / n_trials

    #Smooth the temporal density
    smoothed_consensus = gaussian_filter1d(
        consensus,
        sigma=SMOOTH_SIGMA
    )


    # Detect peaks
    peaks, properties = find_peaks(
        smoothed_consensus,
        prominence=PEAK_PROMINENCE / n_trials,
        distance=PEAK_DISTANCE
    )

    peak_times = time_axis[peaks]

    #Store Output
    all_video_results[(stimulus_name, predictability)] = {
        "stimulus_name": stimulus_name,
        "predictability": predictability,
        "time_axis": time_axis,
        "raw_counts": counts,
        "consensus": consensus,
        "smoothed_consensus": smoothed_consensus,
        "peak_indices": peaks,
        "peak_times": peak_times,
        "peak_heights": smoothed_consensus[peaks],
        "all_boundaries": all_boundaries,
        "n_trials": n_trials
    }
```

```python
# Function of Creating the Plot of the Aggregate Timeline
def plot_consensus_by_condition(video_name):

    plt.figure(figsize=(14, 5))
    
    nob_P = mean_nob.loc[(video_name, "Predictable")]
    nob_U = mean_nob.loc[(video_name, "Unpredictable")]

    for condition in ["Predictable", "Unpredictable"]:

        key = (video_name, condition)

        if key not in all_video_results:
            print(f"Missing result for {video_name} | {condition}")
            continue

        res = all_video_results[key]

        plt.plot(res["time_axis"], res["smoothed_consensus"], linewidth=2, label=f"{condition} consensus")
        plt.scatter(res["peak_times"], res["peak_heights"], s=70, label=f"{condition} peaks")

    
    # Add red shaded window for unpredictable manipulation
    if video_name in unpredictable_windows:
        start, end = unpredictable_windows[video_name]
        plt.axvspan(start, end, color="red", alpha=0.15, zorder=0,label="Unpredictable action window")

    plt.xlabel("Time in Video (seconds)")
    plt.ylabel("Boundary Consensus")
    plt.title(f"{video_name}\n"f"P NoB = {nob_P:.1f} | U NoB = {nob_U:.1f}")
    plt.legend()
    plt.show()
```

```python
plot_consensus_by_condition("baking")
plot_consensus_by_condition("balcony")
plot_consensus_by_condition("bank")
plot_consensus_by_condition("bathroom")
plot_consensus_by_condition("beach")
plot_consensus_by_condition("bedding")
plot_consensus_by_condition("bike")
plot_consensus_by_condition("car")
plot_consensus_by_condition("cleaning")
plot_consensus_by_condition("cereal")
plot_consensus_by_condition("fireplace")
plot_consensus_by_condition("football")
plot_consensus_by_condition("gym")
plot_consensus_by_condition("lamp")
plot_consensus_by_condition("laundry")
plot_consensus_by_condition("mouse")
plot_consensus_by_condition("music")
plot_consensus_by_condition("painting")
plot_consensus_by_condition("party")
plot_consensus_by_condition("poster")
plot_consensus_by_condition("printer")
plot_consensus_by_condition("record")
plot_consensus_by_condition("shopping")
plot_consensus_by_condition("skateboard")
plot_consensus_by_condition("suitcase")
plot_consensus_by_condition("sunbathing")
plot_consensus_by_condition("tea")
plot_consensus_by_condition("tennis")
plot_consensus_by_condition("walking")
plot_consensus_by_condition("whiteboard")
```
